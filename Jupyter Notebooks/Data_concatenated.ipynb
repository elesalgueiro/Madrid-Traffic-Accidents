#### CONCATENATING DATA SETS from 2010 to 2018

**Let´s clean the columns names**

```python
data.columns
```

```python
#Columns names= Delete the spaces and all in lower case
data.columns = data.columns.str.replace(' ', '_').str.lower()
```

```python
#Change the names of columns.
data.rename(columns={'nº_victimas_*': 'victimas','nº':'num_calle','lugar_accidente':'lugar',}, inplace=True)
```
